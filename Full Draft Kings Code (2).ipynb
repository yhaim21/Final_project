```python
total_player_dictionary = {}
url = 'https://fantasydata.com/'
index = 0
for href_link in player_links:
    new_url = url + href_link
    driver.get(new_url)
    soup = BeautifulSoup(driver.page_source, 'lxml')
    tables = soup.find_all('table')
    player = soup.find('div',{'class':'player-title'}).text.strip()
    player = player[:-6]
    player_data = pd.read_html(str(tables[6]))[0]
    player_data = player_data.iloc[0:1]
    player_data.columns = ['Season', 'Team', 'GMS', 'ST', 'PTS', 'REB', 'AST', 'BLK', 'STL', 'FG%', 'FT%', '3P%', 'FTM', '2PM', '3PM', 'TO', 'MIN', 'DD2', 'TD3', 'FPts']
    player_data.loc[:,['ST','PTS','REB','AST','BLK','STL','FTM','2PM','3PM','TO','MIN','FPts']] = player_data.loc[:,['ST','PTS','REB','AST','BLK','STL','FTM','2PM','3PM','TO','MIN','FPts']] / player_data.iloc[0]['GMS']
    player_dictionary = player_data.to_dict()
    ten_most_recent = pd.read_html(str(tables[10]))[0]
    ten_most_recent.columns = ['DATE', 'OPP', 'PTS', 'REB', 'AST', 'BLK', 'STL', 'FG%', 'FT%', '3P%', 'FTM', '2PM', '3PM', 'TO', 'MIN', 'DD2', 'TD3', 'FPTS']
    ten_most_recent.loc[:,['PTS','REB','AST','BLK','STL','FTM','2PM','3PM','TO','FPTS']] = ten_most_recent.loc[:,['PTS','REB','AST','BLK','STL','FTM','2PM','3PM','TO','FPTS']] / 10
    ten_most_recent_dictionary = ten_most_recent.to_dict()
    total_player_dictionary[player] = {}
    total_player_dictionary[player]['opponent'] = fanduel_top_available_adjusted['Abr'][index]
    total_player_dictionary[player]['season_stats'] = player_dictionary
    total_player_dictionary[player]['ten_most_recent'] = ten_most_recent_dictionary
    index += 1
total_player_dictionary
```

```python
new_dictionary = {}
for player_name in total_player_dictionary:
    templist = []
    for cum_metric in total_player_dictionary[player_name]:
        if cum_metric == "opponent":
            templist.append(total_player_dictionary[player_name][cum_metric])
        else:
            for stat in total_player_dictionary[player_name][cum_metric]:
                templist.append(total_player_dictionary[player_name][cum_metric][stat][0])
    new_dictionary[player_name] = templist
new_dictionary
```

```python
player_data_csv = pd.DataFrame.from_dict(new_dictionary, orient='index')
player_data_csv.columns = ['Opp','Season', 'Team', 'GMS', 'ST', 'PTS', 'REB', 'AST', 'BLK', 'STL', 'FG%', 'FT%', '3P%', 'FTM', '2PM', '3PM', 'TO', 'MIN', 'DD2', 'TD3', 'FPts','10DATE', '10OPP', '10PTS', '10REB', '10AST', '10BLK', '10STL', '10FG%', '10FT%', '103P%', '10FTM', '10_2PM', '10_3PM', '10TO', '10MIN', '10DD2', '10TD3', '10FPTS']
player_data_csv
```

```python
player_data_csv.to_csv('player_data_csv.csv', index='false')
```
